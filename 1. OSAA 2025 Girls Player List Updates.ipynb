```python
import json
import csv
import pprint
```

```python
## Clean up some bad names across the OSAA
bad_name_parsing_list=[144173,144187,161966] #these names parse into too many fields
replaced_name_info=[]
corrected_row1=['144173','74631','Ana','Carcano-Blas','Ana Carcano-Blas','11','2']
replaced_name_info.append(corrected_row1)
corrected_row2=['144187','74631','Raquel','Navarro-Karla','Raquel Navarro-Karla','10','2']
replaced_name_info.append(corrected_row2)
corrected_row3=['161966','75425','Kendra','Warren-Galligan','Kendra Warren-Galligan','9','2']
replaced_name_info.append(corrected_row3)

```

```python
## Clean up some bad names from the Pacific Conference
pc_replaced_name_info=[]
corrected_row1=['144173','74631','Ana','Carcano-Blas','Ana Carcano-Blas','10','2']
pc_replaced_name_info.append(corrected_row1)
corrected_row2=['144187','74631','Raquel','Navarro-Karla','Raquel Navarro-Karla','9','2']
pc_replaced_name_info.append(corrected_row2)
```

```python
#List of schools to rate
schools_to_rate_list=[]
schools_to_rate = open('school_league_info.csv')
for line in schools_to_rate:
    fullline=line.strip()
    linelist=fullline.split(',')
    schoolid=linelist[0]
    schools_to_rate_list.append(schoolid)
print(schools_to_rate_list)
```

```python
# Specific Lists for smaller reports
pacific_conference_list=[74631,75069,74639,74645,75289,75359,75744]
roseburg_tourney_list=[74804,124895,75097,124831,75636,75113,75032,75744,75033,75685,74829,7070586]
```

```python
# read file that was pulled from https://api.v2.tennisreporting.com/players
with open('players.json', 'r', encoding='utf-8') as myfile2:
    temp2=myfile2.read()

# parse file
info2 = json.loads(temp2)
```

```python
player_info_list = list()
pc_info_list = list()
roseburg_info_list = list()

for item in info2:
    try:
        player_id=item['id']
        player_school_id=item['schoolId']
        player_first_name=str.strip(item['firstName'])
        player_last_name=str.strip(item['lastName'])
        player_full_name=player_first_name+" "+player_last_name
        if item['grade']=='Graduated':
            player_grade_2025=13
        else:
            player_grade_2025=int(item['grade'])
        player_gender_id=item['genderId']
        if player_grade_2025<13 and player_gender_id==2 and str(player_school_id) in schools_to_rate_list and player_id not in bad_name_parsing_list:
            player_info_list.append([player_id,player_school_id,player_first_name,player_last_name,player_full_name,player_grade_2025,player_gender_id])
        if player_grade_2025<13 and player_gender_id==2 and player_school_id in pacific_conference_list and player_id not in bad_name_parsing_list:
            pc_info_list.append([player_id,player_school_id,player_first_name,player_last_name,player_full_name,player_grade_2025,player_gender_id])
        if player_grade_2025<13 and player_gender_id==2 and player_school_id in roseburg_tourney_list and player_id not in bad_name_parsing_list:
            roseburg_info_list.append([player_id,player_school_id,player_first_name,player_last_name,player_full_name,player_grade_2025,player_gender_id])


    except:
        continue
```

```python
# Use the name replacements above
for x in replaced_name_info:
    player_info_list.append(x)

for x in pc_replaced_name_info:
    pc_info_list.append(x)
```

```python
# Write to 3 versions (all, Pacific Conference, Roseburg Tournament) of the player_info_girls csv files:
with open ('player_info_girls.csv', 'w', newline='\n') as csvfile:
    for item in player_info_list:
        playerwriter=csv.writer(csvfile, delimiter=',',quotechar='|',quoting=csv.QUOTE_MINIMAL)
        playerwriter.writerow(item)

with open ('player_info_girls_pc.csv', 'w', newline='\n') as csvfile:
    for item in pc_info_list:
        playerwriter=csv.writer(csvfile, delimiter=',',quotechar='|',quoting=csv.QUOTE_MINIMAL)
        playerwriter.writerow(item)

with open ('player_info_girls_roseburg.csv', 'w', newline='\n') as csvfile:
    for item in roseburg_info_list:
        playerwriter=csv.writer(csvfile, delimiter=',',quotechar='|',quoting=csv.QUOTE_MINIMAL)
        playerwriter.writerow(item)
```
