### Update the file school_league_info.csv with the schools that do

### If you need school info (for school_league_info.csv), go to:
### https://api.v2.tennisreporting.com/schools

```python
import json
import requests
import csv
import urllib.request, urllib.parse, urllib.error
```

```python
schools_to_rate = open('school_league_info.csv')
for line in schools_to_rate:
    fullline=line.strip()
    linelist=fullline.split(',')
    schoolid=linelist[0]
    schoolstring='https://api.v2.tennisreporting.com/report/school/'+schoolid+'?year=2025&genderId=2'
    #print(schoolstring)
    s = requests.Session()
    login_data = { 'email': '********', 'password': '********' }
    auth = s.post('https://api.v2.tennisreporting.com/auth/login', login_data)
    token = str(json.loads(auth.content)['token'])
    s.headers.update({'token':token})

    matches = s.get(schoolstring)

    #print(matches.text)
    #info = matches.text
    #info = json.loads(matches.text)
    match_json='zz'+schoolid+'match_file.json'
    file_object = open("C:/Users/toddw/Desktop/Python Rating Code and Files/OSAA_2025_Ratings/json_match_files/"+match_json,"w")
    file_object.write(matches.text)
```
